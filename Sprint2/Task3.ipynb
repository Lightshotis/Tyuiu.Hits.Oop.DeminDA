```
using System;

// Интерфейс для книг, которые можно брать
public interface IBorrowable
{
    string Borrow();
    string Return();
}

// Интерфейс для книг, которые можно читать
public interface IReadable
{
    string Read();
}

// Базовый класс Book
public class Book
{
    // Атрибуты
    public string Title { get; set; }
    public string Author { get; set; }
    public int YearOfPublication { get; set; }
    public string Publisher { get; set; }
    public int Pages { get; set; }
    public bool IsAvailable { get; set; }

    // Конструктор
    public Book(string title, string author, int yearOfPublication, string publisher, int pages)
    {
        Title = title;
        Author = author;
        YearOfPublication = yearOfPublication;
        Publisher = publisher;
        Pages = pages;
        IsAvailable = true;
    }

    // Методы
    public virtual string GetInfo()
    {
        return $"Title: {Title}\nAuthor: {Author}\nYear of publication: {YearOfPublication}\nPublisher: {Publisher}\nPages: {Pages}";
    }

    public string Reserve()
    {
        if (IsAvailable)
        {
            IsAvailable = false;
            return "Книга зарезервирована.";
        }
        return "Книга уже зарезервирована.";
    }

    public string ReturnBook()
    {
        IsAvailable = true;
        return "Книга возвращена.";
    }

    public string CheckAvailability()
    {
        return IsAvailable ? "Книга доступна." : "Книга недоступна.";
    }
}

// Производный класс TextBook (Учебник)
public class TextBook : Book, IBorrowable, IReadable
{
    public string Subject { get; set; }
    public string GradeLevel { get; set; }

    // Конструктор
    public TextBook(string title, string author, int yearOfPublication, string publisher, int pages, string subject, string gradeLevel)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Subject = subject;
        GradeLevel = gradeLevel;
    }

    // Переопределенный метод чтения
    public string Read()
    {
        return $"Вы читаете учебник по {Subject}.";
    }

    public string Borrow()
    {
        return "Вы взяли учебник.";
    }

    public string Return()
    {
        return "Вы вернули учебник.";
    }

    // Новый метод
    public string Study()
    {
        return $"Вы изучаете {Subject} на уровне {GradeLevel}.";
    }
}

// Производный класс Fiction (Художественная литература)
public class Fiction : Book, IBorrowable
{
    public string Genre { get; set; }
    public string MainCharacter { get; set; }

    // Конструктор
    public Fiction(string title, string author, int yearOfPublication, string publisher, int pages, string genre, string mainCharacter)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        Genre = genre;
        MainCharacter = mainCharacter;
    }

    // Переопределенный метод взятия книги
    public string Borrow()
    {
        return $"Вы взяли книгу жанра {Genre}.";
    }

    public string Return()
    {
        return "Вы вернули художественную книгу.";
    }

    // Новый метод
    public string DescribeMainCharacter()
    {
        return $"Главный персонаж этой книги — {MainCharacter}.";
    }
}

// Производный класс ScientificLiterature (Научная литература)
public class ScientificLiterature : Book, IReadable
{
    public string FieldOfScience { get; set; }
    public bool IsPeerReviewed { get; set; }

    // Конструктор
    public ScientificLiterature(string title, string author, int yearOfPublication, string publisher, int pages, string fieldOfScience, bool isPeerReviewed)
        : base(title, author, yearOfPublication, publisher, pages)
    {
        FieldOfScience = fieldOfScience;
        IsPeerReviewed = isPeerReviewed;
    }

    // Переопределенный метод получения информации
    public override string GetInfo()
    {
        return $"{base.GetInfo()}\nField Of Science: {FieldOfScience}\nPeer-reviewed: {(IsPeerReviewed ? "Yes" : "No")}";
    }

    // Переопределенный метод чтения
    public string Read()
    {
        return $"Вы читаете научную литературу по {FieldOfScience}.";
    }

    // Новый метод
    public string ConductResearch()
    {
        return $"Вы проводите исследование в области {FieldOfScience}.";
    }
}

        TextBook mathBook = new TextBook("Mathematics for Engineers", "John Doe", 2020, "SciencePress", 350, "Mathematics", "Undergraduate");
        Fiction novel = new Fiction("Mystery of the Old House", "Alice Johnson", 2018, "NovelPress", 250, "Mystery", "Detective Smith");
        ScientificLiterature scienceBook = new ScientificLiterature("Quantum Mechanics", "Albert Einstein", 1925, "PhysicsPress", 400, "Physics", true);

        // Демонстрация методов
        Console.WriteLine(mathBook.Study());                 // Учебник: изучение предмета
        Console.WriteLine(novel.DescribeMainCharacter());    // Художественная книга: описание главного персонажа
        Console.WriteLine(scienceBook.ConductResearch());    // Научная литература: исследование

        // Чтение книги
        Console.WriteLine(mathBook.Read());                 // Чтение учебника
        Console.WriteLine(scienceBook.Read());              // Чтение научной литературы

        // Взятие и возврат книги
        Console.WriteLine(mathBook.Borrow());               // Взять учебник
        Console.WriteLine(mathBook.Return());               // Вернуть учебник

        // Проверка доступности
        Console.WriteLine(novel.CheckAvailability());        // Доступность книги
    

```
